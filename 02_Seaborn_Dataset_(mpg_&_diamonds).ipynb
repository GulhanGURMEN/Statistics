## Seaborn Dataset - "mpg"

 Import **NumPy** and **Seaborn**

```python
import numpy as np
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("mpg")
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.countplot(x="origin", data=df);
```

```python
sns.barplot(x="origin", y="horsepower", data=df);
```

```python
sns.barplot(x="cylinders", y="horsepower", data=df);
```

```python
sns.histplot(x="horsepower", data=df);
```

```python
sns.boxplot(x="horsepower", data=df);
```

```python
sns.scatterplot(x="horsepower", y="weight", data=df);
```

```python
df.corr()
```
